```python
import requests
import pprint
from getpass import getpass

# APIキーと駅名を入力
api_key = getpass('APIキーを入力してください：')
search_station = input('駅名を入力してください：')

# Geocoding APIのURL
url1 = 'https://maps.googleapis.com/maps/api/geocode/json'

# 緯度経度の取得
response = requests.get(
    url1,
    params={
        'address': search_station,  # 駅名を検索
        'key': api_key              # APIキー
    }
)

# レスポンスのJSONを取得
json_data = response.json()


# 緯度・経度を取得
location = json_data['results'][0]['geometry']['location']
latitude = location['lat']
longitude = location['lng']

# Places APIのURL
url2 = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'

# レストラン情報の取得
restaurants_response = requests.get(
    url2,
    params={
        'location': f"{latitude},{longitude}",  # 緯度,経度
        'language': 'ja',                       # 日本語
        'radius': 500,                          # 半径500m
        'type': 'restaurant',                   # レストラン
        'key': api_key                          # APIキー
    }
)

# レスポンスのJSONを取得
restaurants_json_data = restaurants_response.json()


# レストラン情報を表示
for item in restaurants_json_data['results']:
    name = item.get('name', '不明')
    rate = item.get('rating', '評価なし')
    address = item.get('vicinity', '住所不明')
    print("-----------------------------------------------------------------")
    print(f'店名：{name}')
    print(f'評価：{rate}')
    print(f'住所：{address}')
    print("-----------------------------------------------------------------")
```
